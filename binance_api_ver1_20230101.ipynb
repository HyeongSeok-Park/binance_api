```python
import pandas as pd
import ccxt
```

# Binance API data

```python
# 코인 정보 가져오기
binance = ccxt.binanceusdm()  #바이낸스 객체 생성

# 'timeframe' 사용 가능한 변수
#  1분 : 1m 
#  5분 : 5m
#  15분 : 15m
#  30분 : 30m
#  1시간 : 1h
#  6시간 : 6h
#  12시간 : 12h
#  1일 : 1d
#  3일 : 3d
#  1주일 : 1w
#  1달 : 1M

# fetchOHLCV 매서드
# fetchOHLCV ('BTC/USDT', timeframe = '1m', since = 'undefined', limit = 'undefined', params = {})

BTC_data = binance.fetch_ohlcv('BTCUSDT', '1h', since = 1567969200000)  #코인명, 시간간격, 시간대 입력
BTC_data
```

```python
BTC_data = pd.DataFrame(BTC_data, columns = ['datetime','Open','High','Low','Close','Volume'])
BTC_data
```

```python
BTC_data['datetime'] = pd.to_datetime(BTC_data['datetime'], unit='ms')
BTC_data
```
